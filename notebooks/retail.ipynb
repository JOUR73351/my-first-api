```python
limit = '$limit=10000'
parameters = "complaint_type='Consumer Complaint'&descriptor='Retail Store'"
date_filter = "$where=created_date>='2020-01-01T00:00:00'"
url = f'https://data.cityofnewyork.us/resource/erm2-nwe9.json?{date_filter}&{limit}&{parameters}'
unquote(url)
```

```python
response = requests.get(url, headers=headers)
response
```

```python
data = response.json()
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
df['date'] = df['created_date'].astype(str).str[:10]
df.head()
```

```python
by_date = df['date'].value_counts().rename_axis('date').reset_index(name='counts')
```

```python
by_date = by_date.sort_values('date')
```

```python
by_date.to_csv('../output/by_date.csv', index=False)
```
