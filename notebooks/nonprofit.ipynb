```python
import pandas as pd
import requests
```

```python
response = requests.get('https://projects.propublica.org/nonprofits/api/v2/search.json?q=propublica')
```

```python
ein = response.json()['organizations'][0]['ein']
ein
```

```python
fin_data = response.json()['filings_with_data']
df = pd.DataFrame(fin_data)
df
```

```python
results = []
```

```python
eins = ['142007220', '264527097', '464353634', '942434026', '541512177', '591630423', '472474086']
```

```python
for ein in eins:
    response = requests.get(f'https://projects.propublica.org/nonprofits/api/v2/organizations/{ein}.json')
    data = response.json()
    filings = data['filings_with_data']
    for d in filings:
        d['name'] = data['organization']['name']
    results.extend(filings)
```

```python
df = pd.DataFrame(results)
```

```python
pivot_table = pd.pivot_table(df, index=df['tax_prd_yr'], columns=df['name'], values='totrevenue', fill_value='')
```

```python
pivot_table.to_csv('../output/nonprofit.csv', index=False)
```
